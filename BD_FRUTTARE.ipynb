# [dbo].[ComissaoVendedores]

```sql
/****** Object:  View [dbo].[ComissaoVendedores]    Script Date: 02/06/2023 18:46:56 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create view [dbo].[ComissaoVendedores] AS
SELECT v.matricula AS MatriculaVendedor,
       nf.numero AS NumeroNotaFiscal,
       v.percentual_comissao AS PercentualComissao,
       (inf.quantidade * inf.preco * v.percentual_comissao / 100) AS ValorComissao
FROM TABELA_DE_VENDEDORES v
JOIN notas_fiscais nf ON nf.matricula = v.matricula
JOIN itens_notas_fiscais inf ON inf.numero = nf.numero;
GO
```
